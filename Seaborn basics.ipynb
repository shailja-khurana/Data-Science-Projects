# Seaborn
Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

```python
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

# Type of visual representations -Distribution plots-

There are 2 categories of visual representation in Seaborn- 
1. Numerical data - JointPlot, Distplot, Pairplot
2. Categorical data - BarPlot, CountPlot, Voilinplot,Boxplot

### Heatmap and displaying Correlation 
Correlation heatmap uses colored cells ,to show a 2D correlation between numerical fields of dataset.

```python
data.corr()                ## since total_bill,tip and size are numerical fields
```

```python
sns.heatmap(data.corr())          
```

## Jointplot 

A jointplot displays relation between 2 features/numerical variables. 2D representation ,along with central chart displaying correlation.usually scatter,hexbin plot,histogram and 2D density plot.

```python
sns.jointplot(x='tip', y='total_bill', data=data,kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=data,kind='reg')
```

## Pairplot 

This plot pairs all feature with each other to draw scatter or histogram to display their relations.

```python
sns.pairplot(data)  ##below, all numerical fields are shown with graphical relations with each other.
```

```python
## to add one categorical feature to above pairplot, we can do this
sns.pairplot(data, hue='sex')
```

## Dist plot 

It displays distribution of the given column(density distribution).

```python
sns.displot(data['total_bill'])  ##kde=True , bins=20 by default
```

# Categorical feature analysis

## Count plot

```python
sns.countplot(y='smoker',data=data)  ##just like value_counts(),it will display values using graphs
```

```python
sns.countplot(x='day', data=data)
```

## Bar plot

```python
sns.barplot(x='smoker',y='total_bill',data=data)
```

```python
sns.barplot(x='day',y='total_bill',data=data)
```

## Box Plot

```python
sns.boxplot(x='total_bill',y='sex',data=data)
```

```python
sns.boxplot(y='total_bill',x='sex',data=data, palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=data, palette='rainbow')
```

## Violin plot

```python
sns.violinplot(x='total_bill',y='sex',data=data)   ##combination of box plot and distribution density
```
