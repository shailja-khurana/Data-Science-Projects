# Matplotlib 
Matplotlib is one of the most popular Python packages used for data visualization. It is a cross-platform library for making 2D plots from data in arrays. It provides an object-oriented API that helps in embedding plots in applications using Python GUI toolkits such as PyQt, WxPythonotTkinter. It can be used in Python and IPython shells, Jupyter notebook and web application servers also.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(1,20)
y = np.arange(1,20)
```

## Scatter plot

```python
#scatter plot using x and y

plt.scatter(x,y,c='g')          ##c indicates color parameter
#now setting other parameters
plt.xlabel(" x-axis")
plt.ylabel( "y-axis")
plt.title("displying simple 2D graph, with x and y values")

##saving this image
plt.savefig("test.png")
```

```python
## plot function with different parameters

##lets change the values of x nd y
y=x*x

plt.plot(x,y,'r*-',linestyle='dashed',linewidth=2,markersize=12)    ##r*- is styling for the line
plt.xlabel(" x-axis")
plt.ylabel("y-axis")
```

### Creating Subplots

```python
##creating subplots with plot

plt.subplot(2,2,1) ## 2,2,1 means 2 by 2 form of graphs with 2 rows and 2 cols and 1 indicates first graph being plotted
plt.plot(x,y,'b--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'r--')
plt.subplot(2,2,4)
plt.plot(x,y,'y*')
```

```python
#other  forms of graph - like sine form

x = np.arange(0, 4*np.pi,0.1)          ##np.pi =3.14
y = np.sin(x)

plt.plot(x,y,'g--')
```

## Bar plot

```python
x=[2,3,4,5,6,7]
y=[20,40,60,80,90,100]

plt.bar(x,y,color ='g')
plt.xlabel("x-axis")
plt.ylabel('y-axis')
```

## Histogram

```python
x = [23,34,56,12,78,67,56,54,43,32,90,100]

plt.hist(x, bins=10)
plt.title("Histogram")
```

## Box plot

```python
# Box plot ..used mainly for detecting outliers

x = [1,2,12,34,65,78,90,200,300]
plt.boxplot(x, vert=True,patch_artist=True)  ##patch_artist for box color filling and vert means vertical display of chart
plt.title("BoxPlot")
```

```python
#creating list of box plots

list_Data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(list_Data, vert=True, patch_artist=True)
```

# Pie chart
+ 4 parameters need to be specified -
    + labels and size - label for data and size for ratio to be displayed
    + colors - color for each pie
    + explode - how far pie shud be displayed from each other

```python
# Data to plot

data = ['python','ruby','java','html']
size = [400, 330, 200,100]
colors = ['red','green','yellow','blue']
explode = (0.1,0,0,0)

plt.pie(size, explode=explode, labels=data,colors=colors, autopct= "%1.1f%%",shadow=True)
plt.axis('equal')
plt.show()

```
